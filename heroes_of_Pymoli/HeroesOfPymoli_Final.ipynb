## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,8.9,24.9,40.9,46]
groupnames = ['Post-Z', 'Gen-Z','Millenial', 'Pre-Millenial']
df["Age Bins"] = pd.cut(df["Age"], bins, labels=groupnames, include_lowest=True)
df.head()
```

```python
df = pd.DataFrame(df)
df_age_dedup = df.drop_duplicates(subset=['SN']).groupby(["Age Bins"])
df_age = df.drop_duplicates(subset=['SN']).groupby(["Age Bins"])['SN'].agg(['size'])
df_age_summary = pd.DataFrame(df_age)
Total = df_age_summary['size'].sum()

df_age_summary['Percentage of Players'] = df_age_summary['size']/Total
df_age_summary['Percentage of Players'] = df_age_summary['Percentage of Players'].map("{:.0%}".format)
df_age_summary.rename(columns={'size':'Total Count'},inplace=True)
df_age_summary
```
